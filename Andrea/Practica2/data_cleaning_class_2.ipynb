### Llenar datos faltantes para datos categoricos
* Podemos rellenar con el dato especÃ­fico segun el problema
* Generar las siguientes estrategias: 
    *   Usa MODA, si una clase representa mas del 90% de los valores no nulos
    * Usa Una Nueva CategorÃ­a si los datos faltantes son menos de 5% del total de datos y no hay clase dominante.
    * Usar estrategia predictiva si la distribucion de las clases es equilibrada

```python
from collections import Counter
```

```python
class_counts = {}
missing_values = {}

for col in x_cat.columns:
    class_count = Counter(x_cat[col].dropna())  # Excluir valores faltantes al contar clases
    class_counts[col] = class_count
    missing_values[col] = x_cat[col].isnull().sum()  
```

```python
for col, counts in class_counts.items():
    print(f"\nColumna: {col}")
    print(f"Conteos de clases: {counts}")
    print(f"Valores faltantes: {missing_values[col]}")
```

```python
imputer_mode = SimpleImputer(strategy='most_frequent')
imputer_new_category = SimpleImputer(strategy='constant', fill_value="Desconocido")
```

```python
for col, counts in class_counts.items():
    most_common_class, most_common_count = counts.most_common(1)[0]
    total_non_null = sum(counts.values())
 
    
    print(f"\nSugerencia para la columna: {col}")
    
    # Si hay una clase muy dominante (e.g., >90% de los valores), sugiere imputar con la moda
    if most_common_count / total_non_null > 0.9:
        print(f"Sugiere imputar con la moda: '{most_common_class}'")
        mode = imputer_mode.fit_transform(x_train[[col]])
    
        
    # Si no hay una clase dominante pero hay pocos valores faltantes, sugiere crear una nueva categorÃ­a
    elif missing_values[col] / len(x_cat) < 0.05:
        print("Sugiere imputar creando una nueva categorÃ­a ('Desconocido' o 'Faltante').")
        new_category = imputer_new_category.fit_transform(x_train[[col]])
    
    # Si hay una distribuciÃ³n equilibrada entre varias clases, sugiere imputaciÃ³n predictiva
    else:
        print("Sugiere considerar una imputaciÃ³n predictiva o basada en KNN.")
```

### Â¿QuÃ© hacer con datos atÃ­picos?
* Puedes eliminarlos si claramente es un error, pero puedes perder informaciÃ³n y ayuda a eliminar ruido, pero (Â¿quÃ© tal y se trata de una compra inusualmente grande, como un cliente clave?)
* Transformar los Datos
    * Logaritmicas. Util cuando tenemos distribuciones asimÃ©tricas
* Reemplazo de esos valores atÃ­picos
    * Reemplaza con el 95% o 5%
    * Limita los valores a un rango especÃ­fico
* Usa algoritmos menos sensibles a outliers. Los Ã¡rboles de decisiÃ³n o Random Forest y XGBoost, son algortimos menos sensibles a outliers en comparaciÃ³n con SVM y Regresion Lineal.
* CategorizaciÃ³n o Binning. Si tienes Rangos amplios describe datos en "Clases numÃ©ricas".

#### Estrategia Recomendada


    * EliminaciÃ³n de Outliers: Ãštil cuando los outliers son claramente errores o anomalÃ­as sin valor analÃ­tico.
    * TransformaciÃ³n de Datos: Recomendada para datos con distribuciones sesgadas o cuando se busca cumplir con supuestos de normalidad.
    * ImputaciÃ³n de Outliers: Ideal cuando los outliers no son errores, pero se desea limitar su influencia en el anÃ¡lisis.
    * Modelos Robustos: Utilizar si los outliers son comunes o si se sospecha que pueden contener informaciÃ³n valiosa.
    * CategorizaciÃ³n: Considerar en situaciones donde la simplicidad y la reducciÃ³n de la variabilidad son prioritarias.
